```python
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
import joblib

# Encode the target variable first
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Train-test split with encoded target
X_train, X_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)

# Define models with proper parameters
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(random_state=42),
    "GradientBoosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, eval_metric='logloss'),
    "CatBoost": CatBoostClassifier(random_state=42, verbose=0)
}

# Create dictionaries to store results
results_acc = {}
results_reports = {}
trained_models = {}

# Train and evaluate models with pipelines
for name, model in models.items():
    # Create pipeline with scaling
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    
    # Train the pipeline
    pipe.fit(X_train, y_train)
    
    # Make predictions
    y_pred = pipe.predict(X_test)
    
    # Store results
    results_acc[name] = accuracy_score(y_test, y_pred)
    results_reports[name] = classification_report(y_test, y_pred, target_names=label_encoder.classes_)
    trained_models[name] = pipe

# Display results
print("Model Performance Results:")
print("=" * 50)
for name, acc in results_acc.items():
    print(f"--- {name} ---")
    print(f"Accuracy: {acc:.4f}")
    print(results_reports[name])
    print("\n")

# Get best model
best_model_name = max(results_acc, key=results_acc.get)
best_model = trained_models[best_model_name]
best_accuracy = results_acc[best_model_name]

print(f"Best model: {best_model_name} with accuracy {best_accuracy:.4f}")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")
print("Saved best model as best_model.pkl")
print("Saved label encoder as label_encoder.pkl")
```

```python
class SalaryPredictor:
    """
    Advanced Salary Prediction System
    Handles both individual and batch predictions with comprehensive validation
    """
    
    def __init__(self):
        self.model = None
        self.label_encoder = None
        self.feature_options = self._get_feature_options()
        self.load_models()
        
    def _get_feature_options(self) -> Dict[str, list]:
        """Return all feature options for dropdowns"""
        return {
            'workclass': [
                'Private', 'Self-emp-not-inc', 'Self-emp-inc', 'Federal-gov',
                'Local-gov', 'State-gov', 'Without-pay', 'Never-worked', 'NotListed'
            ],
            'education': [
                'Bachelors', 'Some-college', '11th', 'HS-grad', 'Prof-school',
                'Assoc-acdm', 'Assoc-voc', '9th', '7th-8th', '12th',
                'Masters', '1st-4th', '10th', 'Doctorate', '5th-6th', 'Preschool'
            ],
            'marital_status': [
                'Married-civ-spouse', 'Divorced', 'Never-married', 'Separated',
                'Widowed', 'Married-spouse-absent', 'Married-AF-spouse'
            ],
            'occupation': [
                'Tech-support', 'Craft-repair', 'Other-service', 'Sales',
                'Exec-managerial', 'Prof-specialty', 'Handlers-cleaners',
                'Machine-op-inspct', 'Adm-clerical', 'Farming-fishing',
                'Transport-moving', 'Priv-house-serv', 'Protective-serv',
                'Armed-Forces', 'Others'
            ],
            'race': ['White', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other', 'Black'],
            'gender': ['Female', 'Male'],
            'native_country': [
                'United-States', 'Cambodia', 'England', 'Puerto-Rico',
                'Canada', 'Germany', 'Outlying-US(Guam-USVI-etc)', 'India', 'Japan',
                'Greece', 'South', 'China', 'Cuba', 'Iran', 'Honduras',
                'Philippines', 'Italy', 'Poland', 'Jamaica', 'Vietnam',
                'Mexico', 'Portugal', 'Ireland', 'France', 'Dominican-Republic',
                'Laos', 'Ecuador', 'Taiwan', 'Haiti', 'Columbia',
                'Hungary', 'Guatemala', 'Nicaragua', 'Scotland', 'Thailand',
                'Yugoslavia', 'El-Salvador', 'Trinadad&Tobago', 'Peru',
                'Hong', 'Holand-Netherlands'
            ]
        }
    
    def load_models(self) -> None:
        """Load trained model and label encoder with error handling"""
        try:
            self.model = joblib.load("best_model.pkl")
            self.label_encoder = joblib.load("label_encoder.pkl")
            print(" Models loaded successfully!")
            print(f"📊 Model Type: {type(self.model).__name__}")
        except FileNotFoundError as e:
            print(f" Model files not found: {e}")
            print("Please ensure 'best_model.pkl' and 'label_encoder.pkl' are in the current directory")
        except Exception as e:
            print(f" Error loading models: {e}")
            
    def validate_input(self, data: Dict[str, Any]) -> Tuple[bool, str]:
        """Validate user input data"""
        try:
            # Check age range
            if not (16 <= data['age'] <= 100):
                return False, "Age must be between 16 and 100 years"
            
            # Check education number range
            if not (1 <= data['educational_num'] <= 16):
                return False, "Educational level must be between 1 and 16"
            
            # Check hours per week range
            if not (1 <= data['hours_per_week'] <= 99):
                return False, "Hours per week must be between 1 and 99"
            
            # Check capital gains and losses
            if data['capital_gain'] < 0 or data['capital_loss'] < 0:
                return False, "Capital gain and loss cannot be negative"
            
            return True, "Valid input"
            
        except Exception as e:
            return False, f"Validation error: {str(e)}"
    
    def predict_salary(self, input_data: Dict[str, Any]) -> Tuple[Optional[str], Optional[str], Optional[str]]:
        """Make salary prediction for individual input"""
        if not self.model or not self.label_encoder:
            return None, None, "Models not loaded"
        
        try:
            # Validate input
            is_valid, message = self.validate_input(input_data)
            if not is_valid:
                return None, None, message
            
            # Create DataFrame with exact column names expected by model
            input_df = pd.DataFrame({
                'age': [input_data['age']],
                'workclass': [input_data['workclass']],
                'fnlwgt': [input_data['fnlwgt']],
                'educational-num': [input_data['educational_num']],
                'marital-status': [input_data['marital_status']],
                'occupation': [input_data['occupation']],
                'race': [input_data['race']],
                'gender': [input_data['gender']],
                'capital-gain': [input_data['capital_gain']],
                'capital-loss': [input_data['capital_loss']],
                'hours-per-week': [input_data['hours_per_week']],
                'native-country': [input_data['native_country']]
            })
            
            # Make prediction
            prediction_encoded = self.model.predict(input_df)[0]
            prediction_result = self.label_encoder.inverse_transform([prediction_encoded])[0]
            
            # Get confidence if available
            confidence = None
            if hasattr(self.model, 'predict_proba'):
                proba = self.model.predict_proba(input_df)[0]
                confidence = f"{max(proba)*100:.2f}%"
            
            return prediction_result, confidence, None
            
        except Exception as e:
            return None, None, f"Prediction error: {str(e)}"
    
    def predict_batch(self, csv_file_path: str) -> Tuple[Optional[pd.DataFrame], Optional[str]]:
        """Make batch predictions from CSV file"""
        if not self.model or not self.label_encoder:
            return None, "Models not loaded"
        
        try:
            # Read CSV file
            batch_df = pd.read_csv(csv_file_path)
            
            # Validate required columns
            required_columns = [
                'age', 'workclass', 'fnlwgt', 'educational-num', 'marital-status',
                'occupation', 'race', 'gender', 'capital-gain', 'capital-loss',
                'hours-per-week', 'native-country'
            ]
            
            missing_columns = [col for col in required_columns if col not in batch_df.columns]
            if missing_columns:
                return None, f"Missing required columns: {missing_columns}"
            
            # Make predictions
            predictions_encoded = self.model.predict(batch_df[required_columns])
            predictions = self.label_encoder.inverse_transform(predictions_encoded)
            
            # Add predictions to dataframe
            result_df = batch_df.copy()
            result_df['Predicted_Income'] = predictions
            
            return result_df, None
            
        except Exception as e:
            return None, f"Batch prediction error: {str(e)}"

# Initialize the predictor
predictor = SalaryPredictor()
```
